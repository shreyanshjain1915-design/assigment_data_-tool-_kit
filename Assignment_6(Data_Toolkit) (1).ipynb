# Q1. What is NumPy, and why is it widely used in Python?

**Explanation:**

- **NumPy (Numerical Python)** is a powerful open-source Python library mainly used for **numerical computations**.  
- It provides a high-performance **multidimensional array object** called `ndarray`.  
- Operations on arrays in NumPy are **faster and more memory-efficient** compared to regular Python lists.

5. **Memory Efficiency** → Stores large amounts of data in compact form.

---

**Example Code:**

```python
import numpy as np

# Creating a Python list
py_list = [1, 2, 3, 4, 5]

# Creating a NumPy array
np_array = np.array([1, 2, 3, 4, 5])

# Performing operations
print("Python list * 2:", [x * 2 for x in py_list])   # Manual looping
print("NumPy array * 2:", np_array * 2)               # Vectorized operation

# Output:
Python list * 2: [2, 4, 6, 8, 10]
NumPy array * 2: [ 2  4  6  8 10]



# Q2. How does broadcasting work in NumPy?

**Explanation:**

- **Broadcasting** in NumPy allows arrays of **different shapes** to be used together in arithmetic operations.  
- Instead of explicitly replicating smaller arrays, NumPy **automatically stretches** them across the larger array dimensions to perform element-wise operations.  
- This makes the code **shorter, faster, and more memory efficient**.

**→ Broadcasting Rules:**
1. If the arrays differ in dimensions, NumPy compares shapes from **right to left**.
2. Dimensions are **compatible** when:
   - They are equal, or  
   - One of them is `1`.  
3. If not compatible, NumPy throws an **error**.

---

**Example Code 1: Adding a scalar to an array**
```python
import numpy as np

arr = np.array([1, 2, 3, 4])
print("Original Array:", arr)

# Broadcasting: scalar 5 is stretched to match array
result = arr + 5
print("After Broadcasting (arr + 5):", result)

# Output:
Original Array: [1 2 3 4]
After Broadcasting (arr + 5): [ 6  7  8  9]
```

**Example Code 2: Broadcasting with 2D and 1D array**
```python
matrix = np.array([[1, 2, 3],
                   [4, 5, 6]])

vector = np.array([10, 20, 30])

# Broadcasting: vector shape (3,) matches matrix shape (2,3) on last dimension
result = matrix + vector
print("Result:\n", result)

# Output:
Result:
[[11 22 33]
 [14 25 36]]



# Q3. What is a Pandas DataFrame?

# Q4. Explain the use of the `groupby()` method in Pandas

**1. Concept:**
- The `groupby()` method in Pandas is used to **split** a dataset into groups based on one or more columns, then **apply** a function (like sum, mean, count, etc.), and finally **combine** the results.  
- In short: **Split → Apply → Combine**.

**2. Why use `groupby()`?**
- To analyze data at a **category level** rather than row by row.  
- Makes it easy to compute **aggregations** (sum, mean, count, max, min, etc.) for each group.  
- Essential in **data analysis, reporting, and feature engineering**.

---

**3. Syntax:**
```python
df.groupby("column_name")
df.groupby(["col1", "col2"])
```
**4.Example Code : Basic Usage**
```python
import pandas as pd

# Sample dataset
data = {
    "Department": ["IT", "HR", "IT", "Finance", "HR", "Finance"],
    "Employee": ["Ravi", "Anita", "Vikram", "Meera", "Karan", "Sonia"],
    "Salary": [60000, 45000, 70000, 65000, 50000, 62000]

df = pd.DataFrame(data)

```

# Q6. What are the differences between NumPy arrays and Python lists?

**1. Concept:**
- Both **Python lists** and **NumPy arrays** are used to store collections of data, but they are fundamentally different in terms of performance, functionality, and memory usage.  
- **NumPy arrays** are optimized for **numerical and scientific computing**, while **Python lists** are general-purpose containers.

---

**2. Key Differences Between Python Lists and NumPy Arrays**

| Feature                  | Python List                          | NumPy Array                         |
|---------------------------|--------------------------------------|--------------------------------------|
| **Data type**            | Can store multiple data types (heterogeneous) | Stores only one data type (homogeneous) |
| **Performance (speed)**  | Slower (loops in Python)             | Much faster (vectorized C operations) |
| **Memory efficiency**    | Higher memory usage                  | Uses less memory, more compact        |
| **Operations**           | Element-wise operations require loops | Supports element-wise operations directly |
| **Mathematical functions** | Not built-in (need loops/manual coding) | Rich set of mathematical/statistical functions |
| **Best for**             | Small collections, mixed data types   | Large datasets, numerical computations |

---

**3. Example Code: Performance & Operations**

```python
import numpy as np
import time

# Python List
py_list = [1, 2, 3, 4, 5]
# Multiply each element by 2
list_result = [x * 2 for x in py_list]
print("Python List Result:", list_result)

# NumPy Array
np_array = np.array([1, 2, 3, 4, 5])
array_result = np_array * 2
print("NumPy Array Result:", array_result)

# Performance Test
big_list = list(range(1_000_000))
big_array = np.array(big_list)

# Time taken by Python list
start = time.time()
[x * 2 for x in big_list]
end = time.time()
print("List time:", end - start)

# Time taken by NumPy array
start = time.time()
big_array * 2
end = time.time()
print("NumPy array time:", end - start)

# Output:
Python List Result: [2, 4, 6, 8, 10]
NumPy Array Result: [ 2  4  6  8 10]
List time: 0.18 seconds
NumPy array time: 0.01 seconds
```

**4. Summary:**

- Use **Python lists** for small, mixed-type collections.  
- Use **NumPy arrays** for **large-scale numerical computations** where speed and memory efficiency matter.  

→ In short: **Lists are flexible, NumPy arrays are powerful.**

# Q8. What does the term “vectorized operation” mean in NumPy?

**1. Concept:**
- A **vectorized operation** in NumPy means applying a mathematical or logical operation **directly on entire arrays (vectors, matrices, tensors)** without writing explicit Python loops.  
- Internally, NumPy uses **highly optimized C code**, making these operations extremely fast.  

---

**2. Why is vectorization important?**
1. **Speed** → Much faster than Python loops.  
2. **Simplicity** → Cleaner and shorter code.  
3. **Memory efficiency** → Works on arrays directly without intermediate Python objects.  
4. **Consistency** → Operations are applied element-wise automatically.  

---

**3.Example  Code : Without vs With Vectorization**

```python
import numpy as np
import time

# Python loop approach
py_list = [1, 2, 3, 4, 5]
result_loop = []
for x in py_list:
    result_loop.append(x ** 2)  # squaring each element
print("Loop Result:", result_loop)

# NumPy vectorized approach
np_array = np.array([1, 2, 3, 4, 5])
result_vectorized = np_array ** 2
print("Vectorized Result:", result_vectorized)

# Performance comparison (large dataset)
big_array = np.arange(1_000_000)

start = time.time()
[ x**2 for x in big_array ]  # Python list loop
end = time.time()
print("Loop time:", end - start)

start = time.time()
big_array ** 2  # NumPy vectorized
end = time.time()
print("Vectorized time:", end - start)

# Output:
Loop Result: [1, 4, 9, 16, 25]
Vectorized Result: [ 1  4  9 16 25]
Loop time: 0.25 seconds
Vectorized time: 0.01 seconds
```
**4. Quick Summary Table**

| Aspect               | Python Loop                 | NumPy Vectorized Operation |
|-----------------------|-----------------------------|-----------------------------|
| Code Length           | Longer (manual loops)       | Shorter, cleaner            |
| Speed                 | Slow                        | Very fast                   |
| Memory Efficiency     | Less efficient              | More efficient              |
| Best Use Case         | Small, mixed data           | Large-scale numeric data     |

# Q11. What is the role of Seaborn’s `pairplot()` function?

**1. Concept:**
- The `pairplot()` function in Seaborn is used to **visualize pairwise relationships** between multiple numerical variables in a dataset.  
- It automatically creates a **matrix of plots**, showing scatter plots for each pair of variables and histograms (or KDE plots) on the diagonal.  
- Useful for **Exploratory Data Analysis (EDA)** to detect correlations, trends, and distributions quickly.

---

**→ 2. Key Features of `pairplot()`:**
1. **Visualizes relationships** → Helps understand how variables relate to each other.  
2. **Shows distributions** → Histograms or KDE plots on the diagonal for each variable.  
3. **Supports grouping** → Can color points based on a categorical variable (`hue`).  
4. **Quick and efficient** → Creates multiple plots in a single function call.

---

**3. Example Code:**

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

# Sample dataset
data = pd.DataFrame({
    "Math": [88, 92, 80, 89, 100],
    "Science": [84, 94, 70, 88, 90],
    "English": [78, 85, 90, 86, 82],
    "Class": ["A", "A", "B", "B", "A"]
})

# Create a pairplot
sns.pairplot(data, hue="Class", diag_kind="kde")
plt.show()
```
**4. What it does:**
- Off-diagonal plots → Scatter plots showing relationships between pairs of variables (Math vs Science, Math vs English, etc.).

- Diagonal plots → KDE (or histogram) showing distribution of individual variables.

- Hue → Colors points based on categorical variable (Class A/B in this example).

**5. Quick Summary Table**

| Aspect                   | Role of `pairplot()`                          |
|---------------------------|-----------------------------------------------|
| Main Purpose              | Visualize pairwise relationships and distributions |
| Diagonal plots            | Shows variable distribution (hist/KDE)       |
| Off-diagonal plots        | Shows scatter plots for variable pairs       |
| Supports grouping (`hue`) | Color points based on category               |
| Use Case                  | Quick EDA to detect patterns and correlations |


````python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

# Sample dataset
data = pd.DataFrame({
    "Math": [88, 92, 80, 89, 100],
    "Science": [84, 94, 70, 88, 90],
    "English": [78, 85, 90, 86, 82],
    "Class": ["A", "A", "B", "B", "A"]
})

# Create a pairplot
sns.pairplot(data, hue="Class", diag_kind="kde")
plt.show()
````

#Q13. Why is handling missing data important in Pandas?

**1. Concept:**
- **Missing data** refers to **NaN (Not a Number) or null values** in a dataset.  
- Handling missing data is crucial because it can **distort analysis, statistical calculations, and machine learning models** if left unaddressed.

---

**2. Importance of Handling Missing Data:**
1. **Accuracy of Analysis** → Mean, median, correlation, and other calculations can be skewed by missing values.  
2. **Machine Learning Models** → Most ML algorithms cannot handle NaN values directly and may throw errors or produce biased predictions.  
3. **Data Integrity** → Ensures consistency and reliability in reports or dashboards.  
4. **Visualization** → Missing values can mislead plots, graphs, and heatmaps.  
5. **Prevent Errors** → Functions like `sum()`, `mean()`, `groupby()`, or `merge()` may behave unexpectedly with missing values.

---

**3.Example  Code: Handling Missing Data**

```python
import pandas as pd
import numpy as np

# Sample dataset with missing values
data = pd.DataFrame({
    "Name": ["Ravi", "Anita", "Vikram", "Meera"],
    "Age": [25, np.nan, 22, 28],
    "Salary": [60000, 50000, np.nan, 65000]
})

# Detect missing values
print(data.isnull())

# Fill missing values
data_filled = data.fillna({"Age": data["Age"].mean(), "Salary": data["Salary"].mean()})
print("\nData after filling missing values:\n", data_filled)

# Drop rows with missing values
data_dropped = data.dropna()
print("\nData after dropping missing values:\n", data_dropped)

# Output:

2     False  False    True
3     False  False   False

1  Anita  25.0  50000.0

# Q21. Why is NumPy’s array slicing faster than Python’s list slicing?

**1. Concept:**
- Both Python lists and NumPy arrays support slicing, but **NumPy slicing is much faster**.  
- The reason lies in **how data is stored and accessed in memory**.

---

**2. Reasons for Speed Difference:**

| Aspect                     | Python Lists                                  | NumPy Arrays                                     |
|-----------------------------|-----------------------------------------------|--------------------------------------------------|
| **Memory Storage**          | Elements are stored as separate objects scattered in memory. | Elements are stored in **contiguous memory blocks** (like C arrays). |
| **Data Type**               | Can hold mixed data types → less efficient.  | Homogeneous data types → optimized storage.      |
| **Implementation**          | Slicing creates a **new copy** of elements.  | Slicing creates a **view (no-copy)** → just changes the pointer to data. |
| **Underlying Language**     | Pure Python objects → slower overhead.       | Implemented in **C (low-level optimizations)**. |
| **Vectorization**           | Not supported, needs explicit loops.         | Supports vectorized operations on slices directly. |

---

**3. Example  Code:**

```python
import numpy as np
import time

# Python list slicing
py_list = list(range(1000000))
start = time.time()
py_slice = py_list[100:500000]
end = time.time()
print("Python list slicing time:", end - start)

# NumPy array slicing
np_array = np.arange(1000000)
start = time.time()
np_slice = np_array[100:500000]
end = time.time()
print("NumPy array slicing time:", end - start)

# output:
Python list slicing time: 0.0042 sec
NumPy array slicing time: 0.000002 sec

````python
import plotly.express as px
import pandas as pd

# Step 1: Sample data
data = {
    "X": [1, 2, 3, 4, 5],
    "Y": [10, 11, 12, 13, 14],
    "Z": [5, 6, 7, 8, 9],
    "Category": ["A", "B", "A", "B", "A"]  # Optional: color by category
}

df = pd.DataFrame(data)

# Step 2: Create 3D scatter plot
fig = px.scatter_3d(df, x="X", y="Y", z="Z", color="Category", size_max=10, title="3D Scatter Plot Example")

# Step 3: Show plot
fig.show()
````
